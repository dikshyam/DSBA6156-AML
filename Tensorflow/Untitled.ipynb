```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
```

```python
data = pd.read_csv('Sales by Store - Department.xls',sep='\t')
```

```python
data['Month'] = pd.DatetimeIndex(data['Date']).month
```

```python
data['Quarter'] = pd.DatetimeIndex(data['Date']).quarter
```

```python
data['Year'] = pd.DatetimeIndex(data['Date']).year
```

```python
data['Week'] = pd.DatetimeIndex(data['Date']).week
```

```python
data.isna().sum()
```

```python
#monthly_sales = data.groupby(['Store','Dept','Year','Month'])['Weekly_Sales'].sum().reset_index()
```

```python
#monthly_sales['']
```

```python
X = data[['Store', 'Dept', 'IsHoliday', 'Month', 'Quarter',
       'Year']]
Y = data[['Weekly_Sales']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
rfr = RandomForestRegressor(n_estimators = 250, random_state=0)               #RandomForestRegressor
rfr.fit(X_train,y_train)
y_pred=rfr.predict(X_test)
```

```python
print(metrics.mean_absolute_error(y_test, y_pred))
print(metrics.mean_squared_error(y_test, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
accuracy = rfr.score(X_test,y_test)
print('accuracy RandomForestRegressor: ',accuracy*100,'%')
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
data = pd.read_csv('Sales by Store - Department.xls',sep='\t')
```

```python
data['Month'] = pd.DatetimeIndex(data['Date']).month
```

```python
data['Year'] = pd.DatetimeIndex(data['Date']).year
```

```python
data.head(2)
```

```python
monthly = data.groupby(['Year','Month'])['Weekly_Sales'].mean().reset_index()
```

```python
monthly['Date'] = monthly['Year'].map(str)+'-'+monthly['Month'].map(str)+'-01'
```

```python
df = pd.read_csv('air_passengers.csv', parse_dates = ['Month'], index_col = ['Month'])
```

```python
df = monthly[['Date','Weekly_Sales']]
```

```python
df.set_index('Date',inplace=True)
```

```python
plt.subplots(figsize=(40, 20))
plt.xlabel('Date')
plt.ylabel('Avg Sales Across Stores')
plt.plot(df)
```

```python
rolling_mean = df.rolling(window = 12).mean()
rolling_std = df.rolling(window = 12).std()
plt.plot(df, color = 'blue', label = 'Original')
plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean')
plt.plot(rolling_std, color = 'black', label = 'Rolling Std')
plt.legend(loc = 'best')
plt.title('Rolling Mean & Rolling Standard Deviation')
plt.show()
```
